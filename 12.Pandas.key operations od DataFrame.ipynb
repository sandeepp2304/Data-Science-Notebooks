```python
#Key Operations Of Data frame
#concatenate
import pandas as pd
india_weather=pd.DataFrame({
                "city":["mumbai","delhi","banglore"],
                "temparature":[32,45,30],
                "humidity":[80,60,68]
})
india_weather
```

```python
us_weather=pd.DataFrame({
                "city":["Newyork","chicago","orlando"],
                "temparature":[21,14,35],
                "humidity":[68,65,75]
})
us_weather
```

```python
#concatinating the two tables
df=pd.concat([india_weather,us_weather],sort=False)
df
```

```python
#to get continueous data or index values
df=pd.concat([india_weather,us_weather],ignore_index=True,sort=True)
df
```

```python
#to get the data in row wise
df=pd.concat([india_weather,us_weather],axis=1)
df
```

```python
#to get the data in column wise and continueous index
df=pd.concat([india_weather,us_weather],ignore_index=True,axis=0,sort=False)
df
```

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore", 'hyderabad'],
    "temperature": [32,45,30,40]})
temperature_df
```

```python
humidity_df = pd.DataFrame({
    "city": ["delhi","mumbai","banglore"],
    "humidity": [68, 65, 75]})
humidity_df
```

```python
#Merge data
#inner join(it prints only matching of togather)
df=pd.merge(temperature_df,humidity_df,on="city")
df
```

```python
#outer join(it prints other than matching also)
df=pd.merge(temperature_df,humidity_df,on="city",how="outer")
df
```

```python
#merging of india and us weather data
df=pd.merge(india_weather,us_weather,on="city",how="outer")
df
```

```python
#Numerical Indexing
#.loc & .iloc
#to get a particular index we use index
df=pd.DataFrame([1,2,3,4,5,6,7,8,9],index=[44,55,66,77,1,2,3,4,5])
df
```

```python
#it takes index value :loc function 
df.loc[:77]
```

```python
#it takes row number:iloc function uses
df.iloc[:3]
```

```python
#if you insert index value,you get value
df.loc[55]
```

```python
#you insert row value,it gives index
df.iloc[2]
```
