```python
#MATPLOTLIB.PYPLOT
#plot a graph using [2,4,6,8]
C=[2,4,6,8]
import numpy as np
import matplotlib.pyplot as plt
plt.plot(C)
#text the headings on axis
plt.ylabel("NUMBERS")
plt.xlabel("INDICES")
plt.title("MY PLOT")
plt.grid()
plt.style.use('fast')
plt.show()#by default all the values are taken on y-axis
```

```python
import numpy as np 
from matplotlib import pyplot as plt 

x = np.arange(0,10) 
y = x ^ 2 
z = x ^ 3
t = x ^ 4 
# Labeling the Axes and Title
plt.title("Graph Drawing") 
plt.xlabel("Time") 
plt.ylabel("Distance") 
plt.plot(x,y)
#Annotate
plt.annotate(xy=[2,1], s='Second Entry') 
plt.annotate(xy=[4,6], s='Third Entry') 
# Adding Legends
plt.plot(x,z)
plt.plot(x,t)
plt.legend(['Race1', 'Race2','Race3'], loc=4) 

```

# bar graphs

```python
from matplotlib import pyplot as plt 
x = [5,8,10] 
y = [12,16,6]  

x2 = [6,9,11] 
y2 = [6,15,7] 
plt.bar(x, y, align = 'center') 
plt.bar(x2, y2, color = 'g', align = 'center') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')
plt.legend(["good","bad"])
plt.show()
```

# draw a graph

```python
#draw a graph
import numpy as np
t=np.arange(0.,5.,0.2)
plt.plot(t,t**2,"b--",label="^2",linewidth=5)#it takes dotted line
plt.plot(t,t**2.2,"rs",label="^2.2")#it takes square shape
plt.plot(t,t**2.5,"g",label="^2.5")#by default it takes straight line
plt.plot(t,t**2.7,"yo",label="^2.7")#it takes circle shape
plt.grid()#it allows the grid lines
plt.title("label graph")
plt.legend()#it enables the respective figures of the graph
plt.show()
```

# multiple graphs

```python
plt.figure(1)#first figure
plt.subplot(211)#sub plot in the first figure
plt.plot([2,1,2])
plt.subplot(212)#the second subplot in the first figure
plt.plot([4,5,6])

plt.figure(2)#a second figure
plt.plot([4,5,6])#creates a subplot(111) by default

plt.figure(1)#first figure current; subplot(212) still current
plt.subplot(211)#sub plot 211 title
plt.title("easy 1,2,3")
plt.show()
```
