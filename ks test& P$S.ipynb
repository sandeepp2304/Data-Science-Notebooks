# KS TEST

```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt


#generate a gaussian r.v X
x = stats.norm.rvs(size=1000);
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5)
plt.show()
```

```python
stats.kstest(x, 'norm')
```

```python
# Y ~ Continous Uniform Distribution(0,1)
y = np.random.uniform(0,1,10000);
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
```

# PROBABILITY&STATISTICS

```python
#generate random variables
import numpy as np
import pylab
import scipy.stats as status
#N(0,1)
std_normal=np.random.normal(loc=1,scale=1,size=1000)
#0 to 100th percentiles of std deviation
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
#generate 100 samples from N(20,5)
measurments=np.random.normal(loc=20,scale=5,size=100)
status.probplot(measurments,dist="norm",plot=pylab)
pylab.show()
```

```python
#generate 100 samples from N(20,5)
measurments=np.random.uniform(low=-1,high=1,size=100)
status.probplot(measurments,dist="norm",plot=pylab)
pylab.show()
```

```python
#random number generator
import random
print(random.random())
#ranges from 0 to 1
from sklearn import datasets
iris=datasets.load_iris()
d=iris.data
d
```

```python
#sample 30 points randomly from the 150 points
n=150;
m=30;
p=m/n;#=0.2
sample_data=[];
for i in range(0,n):
    if random.random()<=p:
        sample_data.append(d[i,:])
len(sample_data)
print(sample_data)
```

# box cox 

```python
from scipy import stats
import matplotlib.pyplot as plt
#we generate some random variable from non normal distribution and make probability plot
fig=plt.figure()
ax1=fig.add_subplot(211)
x=stats.loggamma.rvs(5,size=500)+5
prob=stats.probplot(x,dist=stats.norm,plot=ax1)
ax1.set_xlabel(" ")
ax1.set_title("probability against normal distribution")
print("\n")

ax2=fig.add_subplot(212)
xt,_=stats.boxcox(x)
prob=stats.probplot(xt,dist=stats.norm,plot=ax2)
ax2.set_title("probability after boxcox transformation")
```
