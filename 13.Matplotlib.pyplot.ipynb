```python
#draw a graph
import numpy as np
t=np.arange(0.,5.,0.2)
plt.plot(t,t**2,"b--",label="^2")#it takes dotted line
plt.plot(t,t**2.2,"rs",label="^2.2")#it takes square shape
plt.plot(t,t**2.5,"g",label="^2.5")#by default it takes straight line
plt.plot(t,t**2.7,"yo",label="^2.7")#it takes circle shape
plt.grid()#it allows the grid lines
plt.legend()#it enables the respective figures of the graph
plt.show()
```

```python
#working with multipple figures and axis
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
t3 = np.arange(0.0, 5.0, 0.03)

plt.figure(1)
plt.subplot(311)#subplot(row,column,figure(1) divided in to two parts,why because 2 rows,1 indicates first division)
plt.grid()
plt.plot(t1, f(t1), 'b-')
plt.subplot(312)#subplot(row,column,figure(1) divided in to two parts,why because 2 rows,1 indicates first division)
plt.grid()
plt.plot(t2, f(t2), 'ro-')
plt.subplot(313)#subplot(row,column,figure(1) divided in to two parts,why because 2 rows,1 indicates first division)
plt.grid()
plt.plot(t3, f(t3), 'g--')
```
